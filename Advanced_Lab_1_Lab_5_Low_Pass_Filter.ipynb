```python
import matplotlib.pyplot as plt
import numpy as np

path = "/content/falstadlowpassfilterdata.csv" #data from Falstad simulator
data = np.genfromtxt(path, delimiter=',', names='frequency, voltage')

freq = data['frequency']
v_out = data['voltage']

mag = 10*np.log10(v_out/5)
cutoff = 72.34

plt.figure(figsize = (12, 6))
plt.subplot(121)
plt.title('Frequency vs Voltage for a Simulated RC Low Pass Filter')
plt.plot(freq, v_out)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Voltage (V)')

plt.subplot(122)
plt.plot(freq, mag)
plt.xscale('log')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
plt.title('Bode Plot for a Simulated RC Low Pass Filter')
plt.plot(freq[3], mag[3], 'o', color='tab:blue')

plt.annotate("Cutoff Frequency = 72.34 Hz \n mag = -3 dB", # this is the text
xy=(freq[3], mag[3]), # these are the coordinates to position the label
textcoords="offset points", # how to position the text
xytext=(-50,-50), # distance from text to points (x,y)
ha='center',
va='bottom',
bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.2),
arrowprops=dict(arrowstyle = '-|>', connectionstyle='arc3, rad=0'))
plt.tight_layout()

plt.show()
```
