```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.fft import fft, ifft, fftfreq

path = "/content/Marlene McKinney Advanced Lab 1 Lab 4 Data.csv"
data = np.genfromtxt(path, delimiter=',', names=['times','freq'])

time = data['times']*0.001
frequency = data['freq']

X = fft(frequency)
N = len(X)
dt = 0.001
T = dt
freq = fftfreq(N, T)[:N//2]

plt.figure(figsize = (12, 6))
plt.subplot(121)
plt.title('Signal Frequencies')
plt.stem(freq, np.abs(X)[:N//2], 'b', \
         markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('FFT Amplitude |X(freq)|')
plt.xlim(0, 10)
plt.ylim(0, 1.5e6)

plt.subplot(122)
plt.title('Amplitude vs. Time')
plt.plot(time, np.real(ifft(X)), 'r')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```
