```python
import matplotlib.pyplot as plt
import numpy as np

path = "/content/lab5advlab1data.csv"
data = np.genfromtxt(path, delimiter=',', names='time, raw, avg1, avg5, avg10')

t = data['time']*0.001
raw = data['avg1']
avg5 = data['avg5']
avg10 = data['avg10']

fig, ax = plt.subplots(figsize=(6,3), dpi=150)

ax.plot(t, raw, color='black', label="Raw Data")
ax.plot(t, avg5, color='red', label="5 Reading Smoothing")
ax.plot(t, avg10, color='steelblue', label="10 Reading Smoothing")

ax.set_xlabel('Time (s)')
ax.set_ylabel('Signal')
ax.set_title('LED Signal and Noise Attenuation')
ax.grid()
plt.legend()
plt.show()
```
